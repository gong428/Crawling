```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time

from dotenv import load_dotenv
import os

load_dotenv()


# 유튜브 API 키 설정
api_key = os.getenv('YOUTUBE_API')

def search_youtube(keyword, max_videos):
    # Setup the webdriver (Chrome)
    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service)
    
    # Open YouTube
    driver.get("https://www.youtube.com")
    
    # Find the search bar and input the keyword
    search_box = driver.find_element(By.NAME, "search_query")
    search_box.send_keys(keyword)
    search_box.send_keys(Keys.RETURN)
    
    time.sleep(3)  # Wait for the page to load
    
    results = []
    scroll_pause_time = 2

    while len(results) < max_videos:
        # Find video titles, links, upload dates, and view counts
        videos = driver.find_elements(By.XPATH, '//*[@id="video-title"]')
        upload_dates = driver.find_elements(By.XPATH, '//*[@id="metadata-line"]/span[2]')
        view_counts = driver.find_elements(By.XPATH, '//*[@id="metadata-line"]/span[1]')
        
        for video, upload_date, view_count in zip(videos, upload_dates, view_counts):
            title = video.get_attribute("title")
            link = video.get_attribute("href")
            upload_date_text = upload_date.text
            view_count_text = view_count.text
            if title and link and (title, link, upload_date_text, view_count_text) not in results:
                results.append((title, link, upload_date_text, view_count_text))
                
                if len(results) >= max_videos:
                    break

        # Print progress
        print(f"Collected {len(results)} videos so far...")

        # Scroll to the bottom of the page
        driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
        time.sleep(scroll_pause_time)  # Wait for the new videos to load

    driver.quit()
    return results

def save_to_csv(keyword, results):
    df = pd.DataFrame(results, columns=["Title", "Link", "Upload Date", "View Count"])
    filename = f"{keyword}_youtube_result.csv"
    df.to_csv(filename, index=False, encoding='utf-8-sig')
    print(f"Results saved to {filename}")

def main():
    keyword = input("Enter the search keyword: ")
    max_videos = int(input("Enter the number of videos to search: "))
    results = search_youtube(keyword, max_videos)
    
    for idx, (title, link, upload_date, view_count) in enumerate(results, start=1):
        print(f"{idx}. {title}\n   {link}\n   {upload_date}\n   {view_count}\n")
    
    save_to_csv(keyword, results)

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
from datetime import datetime, timedelta
import re

def convert_relative_date_to_absolute(relative_date: str, base_date: datetime) -> datetime:
    """
    상대 날짜 문자열을 기준 날짜를 기준으로 절대 날짜로 변환하는 함수
    """
    num = int(re.findall(r'\d+', relative_date)[0])
    if '년' in relative_date:
        return base_date - timedelta(days=num*365)
    elif '개월' in relative_date:
        return base_date - timedelta(days=num*30)
    elif '주' in relative_date:
        return base_date - timedelta(weeks=num)
    elif '일' in relative_date:
        return base_date - timedelta(days=num)
    elif '시간' in relative_date or '분' in relative_date:
        return base_date - timedelta(hours=num)
    else:
        raise ValueError("지원하지 않는 형식입니다")

def calculate_absolute_dates(df: pd.DataFrame, base_year: int, base_month: int) -> pd.DataFrame:
    """
    데이터프레임의 상대 날짜를 기준 날짜를 기준으로 절대 날짜로 변환하는 함수
    """
    # 기준 날짜 생성
    base_date = datetime(base_year, base_month, 1)
    
    # 상대 날짜를 절대 날짜로 변환
    df['Absolute Upload Date'] = df['Upload Date'].apply(lambda x: convert_relative_date_to_absolute(x, base_date))
    
    return df

# 사용자 입력: 기준 연도와 월
base_year = 2024
base_month = 7

# 절대 날짜로 변환
result_df = calculate_absolute_dates(df, base_year, base_month)
result_df
```
